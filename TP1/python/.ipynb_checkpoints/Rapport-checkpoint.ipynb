```python
%matplotlib inline
import BayesNetwork as bn
import FactorGraph as fg
import IPythonToolbox as ipy
from tabulate import tabulate
```

```python
ex1 = bn.Network()
V = ex1.createNode("V", 0.80)
W = ex1.createNode("W", 0.64)
Y = ex1.createNode("Y", [0.8,0.01,0.8,0.01], parents=[V,W])
ex1.drawGraph()
```

```python
bnet = bn.Network()
C = bnet.createNode("C", 0.001)
T = bnet.createNode("T", 0.002)
A = bnet.createNode("A", [0.95,0.94,0.29,0.001], parents=[T,C])
M = bnet.createNode("M", [0.90,0.05], parents=A)
J = bnet.createNode("J", [0.70,0.01], parents=A)
bnet.drawGraph()
```

```python
bnet.drawDistribution()
```

```python
graph = fg.graphFromBayesNet(bnet)
graph.drawGraph()
```

```python
knowing = [(M==True)&(J==False), (M==False)&(J==True), (M==True)&(J==True), (M==False)&(J==False), M==True, J==True]
probas = [[bnet.p(C==True,k)*100, graph.p(C==True,k)*100] for k in knowing]
pNames = ["p( C=True | "+k.printVars()+")" for k in knowing]
title = "Calcul de Probabilitées par le réseau de Bayes et l'algorithme de Sum-Product"
ipy.printTable(probas, vHeader=pNames, title=title, hHeader=[ 'Bayes','SumProduct'], roundFloat=3, suffix=' %')
```

```python
graph.addObserved(C==False)
```
